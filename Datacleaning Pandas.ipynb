```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df
```

# FOLLOWING CHANGED MADE IN THE COLUMN NAMED FLIGHT NUMBER:
## 1. MISSING VALUES ADDED
## 2. DATA TYPE CHANGED TO INTEGER FROM FLOAT

```python
df["FlightNumber"].iloc[1] = 10055
df["FlightNumber"].iloc[3] = 10075
```

```python
def fun_int(num):
    return int(num)
```

```python
df["FlightNumber"] = df["FlightNumber"].apply(fun_int)
```

## SPLITTED THE FROM_TO COLUMN INTO TWO SEPERATE COLUMNS

```python
def fun_splt_0(string):
    return string.split("_")[0]

def fun_splt_1(string):
    return string.split("_")[1]
```

```python
df["From_To"].apply(fun_splt_0)
```

```python
df["From_To"].apply(fun_splt_1)
```

```python
df["From"] = df["From_To"].apply(fun_splt_0)
```

```python
df["To"] = df["From_To"].apply(fun_splt_1)
```

# DELETED FROM_TO COLUMN FROM THE DATA FRAME

```python
del df["From_To"]
```

## STANDARDISE THE STRING SO THAT ONLY THE FIRST LETTER IS IN UPPERCASE

```python
def fun_stand(string):
    return string.capitalize()
```

```python
df["From"] = df["From"].apply(fun_stand)
```

```python
df["To"] = df["To"].apply(fun_stand)
```
